```python
data_ps['Reviews'] = data_ps['Reviews'].replace({'3.0M':'3000000'})
data_ps['Reviews']= data_ps['Reviews'].astype(str).astype(int)
data_ps[data_ps['Reviews']>3000]
```

```python
data_ps.info()
```

```python
data_ps['Rating'].corr(data_ps['Reviews'])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
dt1 = data_ps[data_ps['Reviews']>1000000]
#dt1['Rating']
ax.scatter(dt1['Rating'], dt1['Reviews'])
plt.xlim([3, 5])
plt.ylim([1000000, 90000000])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(16, 6))
dt_g = dt1['Genres'].value_counts()
print(dt_g.index)
dt2 = dt1[dt1['Genres'].isin(dt_g[dt_g>20].index)]
#ax1.scatter(dt1['Genres'], dt1['Rating'])
ax1.scatter(dt2['Genres'], dt2['Reviews'])
```

```python
fig, ax2 = plt.subplots(figsize=(16, 6))
dt_c = dt1['Category'].value_counts()
print(dt_c)
dt3 = dt1[dt1['Category'].isin(dt_c[dt_c>15].index)]
ax2.scatter(dt3['Category'], dt3['Reviews'])
```
