```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_data.csv')
df = df.dropna()
df
```

```python
df['Month'] = df['Order Date'].str[0:2]
df.drop(df.loc[df['Month']=='Or'].index, inplace=True)
#df['Month'] = df['Month'].replace({'Or':'00'})
df['Month'] = df['Month'].astype(str).astype(int)
df.head()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df['Sales'] = df['Quantity Ordered']* df['Price Each']
df.info()
```

```python
df1 = df.groupby('Month').sum()
```

```python
plt.bar(df1.index, df1['Sales'])
plt.xticks(df1.index)
plt.ylabel('Sales($)')
plt.xlabel('Month')
plt.show()
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1])
df.head()
```

```python
df2 = df.groupby('City').sum()
```

```python
#plt.figure(figsize=[12,8])
plt.bar(df2.index, df2['Sales'])
plt.xticks(df2.index, rotation='vertical')
plt.ylabel('Sales($)')
plt.xlabel('City')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df
```

```python
df3 = df.groupby('Hour').count()
```

```python
plt.plot(df3.index, df3)
plt.xticks(df3.index)
plt.grid()
plt.show()
```

```python
df4 = df[df['Order ID'].duplicated(keep=False)]
df4['Grouped'] = df4.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df4 = df4[['Order ID', 'Grouped']].drop_duplicates()
df4
```

```python
from  itertools import combinations
from collections import Counter

count = Counter()

for row in df4['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

```python
product_group = df.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df4 in product_group]
plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical')
plt.show()
```

```python
price = df.groupby('Product').mean()['Price Each']

print(price)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='blue')
ax2.plot(products, price, 'g--')

ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered')
ax2.set_ylabel('price')
ax1.set_xticklabels(products, rotation='vertical')

plt.show()
```

```python
df5 = df.groupby(['Month', 'Product']).sum()
df5 = df5.loc[1]
plt.bar(df5.index, df5['Quantity Ordered'])
plt.xticks(df5.index, rotation='vertical')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
plt.style.use('seaborn')
df5 = df.pivot_table(index='Month', columns='Product', values='Quantity Ordered', aggfunc='sum')
month = df5.index
df5.plot(ax=ax)
plt.xticks(month)
plt.show()
```
